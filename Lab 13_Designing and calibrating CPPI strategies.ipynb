### Interactive CPPI Simulation (Monte Carlo)

```python
cppi_controls = widgets.interactive(rk.show_cppi,
                                   n_scenarios=widgets.IntSlider(min=1, max=1000, step=1, value=50, description='# Scenarios'),
                                    mu=widgets.FloatSlider(min=0.0, max=0.2, step=0.01, value=0.07, description='Mean'),
                                    sigma=widgets.FloatSlider(min=0.0, max=0.50, step=0.05, value=0.15, description='Volatility'),
                                    floor=widgets.FloatSlider(min=0.0, max=2.0, step=0.1, value=0, description='Floor'),
                                    m=widgets.FloatSlider(min=1.0, max=5.0, step=0.5, value=3.0, description='Multiplier'),
                                    risk_free_rate=widgets.FloatSlider(min=0.0, max=0.05, step=0.01, value=0.03, description='Risk-free Rate'),
                                    steps_per_year=widgets.IntSlider(min=1, max=12, step=1, value=12, description='Rebals/year'),
                                    y_max=widgets.IntSlider(min=0, max=100, step=1, value=100, description='Zoom Y axis')
                                   )
```
